```python
from pyecharts import Scatter
```

```python
extra_name = list(cleandf['title'])
cleandf['years'] = cleandf['years'].astype(int)
extra_data = list(cleandf['years'])
def custom_formatter(params):
    return params.value[3]

scatter = Scatter("top100")
scatter.add("top100", 
            cleandf['tomatometer'], 
            cleandf['reviews_counted'],
            extra_data=extra_data,
            extra_name=extra_name, 
            xaxis_min=90,
            xaxis_name="Tomato Scores",
            yaxis_name="Review Numbers",
            yaxis_name_gap=40,
            visual_dimension=2,
            visual_orient="horizontal",
            visual_type="size",
            visual_range=[17000, 28000],
            visual_text_color="#000",
            tooltip_formatter=custom_formatter)
scatter.show_config()
scatter.render('ratings.html')
```

```python
from pyecharts import Pie
```

```python
def round_down(num):
    return num - (num%10)

yearzone = []

for i in cleandf['years']:
    yearzone.append(round_down(i))
```

```python
cleandf['yearsr'] = yearzone
decades = dict(cleandf['yearsr'].value_counts())
decade = list(decades.keys())
count = list(decades.values())
```

```python
pie = Pie("top100 year span")
pie.add("top100 year span", decade, count, xaxis_min=1920)
pie.show_config()
pie.render('yearspan.html')
```

```python
cleandf['aud_score_percent'] = cleandf['average_rating'] * 20
```

```python
cleandf['critic-audience'] = cleandf['tomatometer'] - cleandf['aud_score_percent']
```

```python
extra_name = list(cleandf['title'])
extra_data = list(cleandf['years'])
cleandf['number_of_ratings'] = cleandf['number_of_ratings'].astype(int)
cleandf['reviews_counted']=cleandf['reviews_counted'].astype(int)
def custom_formatter(params):
    return params.value[3]

scatter = Scatter("top100")
scatter.add("top100", 
            cleandf['reviews_counted'], 
            cleandf['number_of_ratings'],
            extra_data=extra_data,
            extra_name=extra_name, 
            #xaxis_max=35,
            xaxis_name="Differences between audience and critics",
            yaxis_name="number of audience rating",
            yaxis_name_gap=70,
            
            visual_dimension=2,
            visual_orient="horizontal",
            visual_type="size",
            visual_range=[17000, 28000],
            visual_text_color="#000",
            tooltip_formatter=custom_formatter)
scatter.show_config()
scatter.render('difference between critic and audience.html')
```
