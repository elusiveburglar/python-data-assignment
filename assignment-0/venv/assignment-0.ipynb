```python
stopwords = ['the', 'a', 'an', 'it', 'that', 'to', 'its', 'of' , 'for']
for word in list(words):  # iterating on a copy since removing will mess things up
    if word in stopwords:
        words.remove(word)
```

```python
delete = ['', 'this', 'on', 'in', 'has', 'that', 'is', 'are', 'they', 'be', '[', ']', 'what', 'and','And', 'I', 'my', 'By', 'by',
          'at', 'What', 'The','’', 'which', 'will', 's', 'we', 'so', 'all', 'This', 'who', 'than', 'but', 'with', 'as',
          'from', 'also', 'have', 'would', 'not','was','He', 'been','while','or','His', 'his', 'can', 'had','']
words[:] = [elem for elem in words if elem not in delete]
```

```python
newords = [elem for elem in words]
```

```python
newords
```

```python
from collections import Counter
d = Counter(newords)
```

```python
s = [(k, d[k]) for k in sorted(d, key=d.get, reverse=True)]
for k, v in s:
    k, v
```
